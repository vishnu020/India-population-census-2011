```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)


df=pd.read_csv(r"C:\Users\HP\Desktop\python_project\movies.csv")
pd.options.mode.chained_assignment = None
```

```python
for col in df.columns:
   pct_missing= np.mean(df[col].isnull())
   print("{}-{}%".format(col,pct_missing))
```

```python
df.dtypes
```

```python
df['gross']=df['gross'].fillna(0)
```

```python
df['budget']=df['budget'].fillna(0)
```

```python
df['gross']=df['gross'].astype('int64')
df['budget']=df['budget'].astype('int64')
```

```python
df
```

```python
df.head()
```

```python
df=df.sort_values('gross',inplace=False,ascending=False)
```

```python
pd.set_option('display.max_rows',None)
```

```python
df['company']=df['company'].drop_duplicates()
```

```python
df.head()
```

```python
plt.scatter(x=df['budget'],y=df['gross'])
plt.title("budget vs gross")
plt.xlabel("grossearning")
plt.ylabel("budet of movie")
plt.show()
```

```python
correlation_metrix=df.corr(method='pearson')
sns.heatmap(correlation_metrix,annot=True)
plt.title("correlation_metrix")
plt.xlabel("movie_features")
plt.ylabel("movie_features")
plt.show()
```

```python
df_numerized=df
```

```python
for col_name in df_numerized.columns:
    if df_numerized[col_name].dtype=='object':
        df_numerized[col_name]=df_numerized[col_name].astype("category")
        df_numerized[col_name]=df_numerized[col_name].cat.codes
   
df_numerized.head()
        
```

```python
correlation_metrix=df_numerized.corr(method='pearson')
sns.heatmap(correlation_metrix,annot=True)
plt.title("correlation_metrix")
plt.xlabel("movie_features")
plt.ylabel("movie_features")
plt.show()
```
